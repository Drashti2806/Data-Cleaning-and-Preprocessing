# Level 1 : Task 2 Exploratory Data Analysis

```python
import pandas as pd

df = pd.read_csv("house_prediction.csv")
df.head()  
```

```python
summary_stats = df.describe().T
summary_stats['median'] = df.median()
summary_stats['mode'] = df.mode().iloc[0]

summary_stats
```

```python
import matplotlib.pyplot as plt

df.hist(figsize=(15,10), bins=20, edgecolor='white')
plt.tight_layout()
plt.show()
    
```

```python
import seaborn as sns

features = ['rooms', 'low_income', 'crime_rate', 'price']

for feature in features:
    plt.figure(figsize=(6,4))
    sns.boxplot(y=df[feature])
    plt.title(f'Boxplot of {feature}')
    plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12,8))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Matrix")
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

top_features = ['rooms', 'low_income', 'student_teacher_ratio', 'crime_rate']

# Plot scatterplots
for feature in top_features:
    plt.figure(figsize=(6,4))
    sns.scatterplot(x=df[feature], y=df['price'])
    plt.title(f'Price vs {feature}')
    plt.xlabel(feature)
    plt.ylabel('Price')
    plt.show()
```
